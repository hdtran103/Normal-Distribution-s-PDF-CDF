# Imports 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

# Load Data
- I've got this dataset with info about how tall people are, and I'm specifically looking at the heights of girls. I want to figure out some cool stuff from it.

```python
df = pd.read_csv('Data/humanheightdata.csv')
males = df.loc[ df['Gender'] == 'Male'].copy()
females = df.loc[ df['Gender'] == 'Female'].copy()
```

- For more reproducible code, we will define a variable for our data.

```python
data = females['Height']
```

```python
low_end = 55.0
high_end = 56.0
```

- I made this graph called a histogram. 
- It shows me how many girls have heights in different ranges. 
- See those black lines? They're like the borders of a special range I care about, from 55.0 inches to 56.0 inches. 
- That yellow part? That's the range I'm focusing on.
- The curvy line is like a smooth version of the graph
  - which helps me see if there's a peak height in that yellow zone.

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(x=data, stat='density',kde=True)
ax.set_title('Distribution of Female Height')
ax.axvline(low_end, color= 'black')
ax.axvline(high_end, color ='black')
ax.axvspan(low_end, high_end, color ='yellow', zorder = 0);
```

```python
mean = np.mean(data)
std = np.std(data)
min = np.min(data)
max = np.max(data)
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(x=data, stat='density',kde=True,cumulative=False)
ax.set_title('Distribution of Female Height')
ax.axvline(mean,color='k',ls='--');
```

### Probability of Less than or Greater than a Value

Instead of finding the probability of a specific range, we may instead want to know the probability of a value being less than or greater than a particular value.

For example, if we wanted to calculate the probability of the height being less than 60 inches, we don't need to subtract.

```python
stats.norm.cdf(60, loc=mean,scale=std)
```
